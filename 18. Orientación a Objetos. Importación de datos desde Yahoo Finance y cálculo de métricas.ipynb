# Orientación a Objetos. Importación y análisis

Vamos a crear un sistema que nos permita descargar los datos directamente desde internet, creando una librería.

Importamos librerías, si no está instalado yfinance previamente habría que instalarlo, como en ejercicios anteriores ya lo he instalado tengo la librería y solo tengo que llamarla.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
plt.style.use("seaborn")
```

```python
class AnalisisFinanciero():
    ''' Class para realizar análisis empresas cotizadas.

     Atributos
    ===========
    ticker: str
        ticker es el símbolo que utiliza Yahoo para publicación de cotizaciones
    start: str
        fecha inicial para extraer históricos de cotizaciones
    end: str
        fecha final para extraer historicos de cotizaciones

     Métodos
    =========
    get_data:
        extracción de histórico de cotizaciones de la página de Yahoo Finance y preparación de datos para posterior análisis
    log_ret:
        cálculo de retornos de las acciones
    plot_precios:
        gráfico de precios
    plot_ret:
        gráfico de rentabilidades dos opciones: como serie de tiempo ("ts") o histograma ("hist")
    set_ticker:
        selección nuevo ticker (símbolo)
    mean_ret:
        cálculo de media de las rentabilidades
    std_ret:
        cálculo de desviación típica de las rentabilidades (riesgo)
    anual_perf:
        cálculo anualizado de rentabilidades y riesgo
    '''
    
    def __init__(self, ticker, start, end):
        self.ticker = ticker
        self.start = start
        self.end = end
        self.get_data()
        self.log_ret()
    
    def __repr__(self): 
        return "AnalisisFinanciero(ticker = {}, start = {}, end = {})".format(self.ticker, 
                                                                               self.start, self.end)
    def get_data(self):
        ''' Extracción de histórico de cotizaciones de la página de Yahoo Finance y preparación de datos para posterior análisis
        '''
        data = yf.download(self.ticker, self.start, self.end).Close.to_frame()
        data.rename(columns = {"Close":"Precio"}, inplace = True)
        self.data = data
        
    def log_ret(self):
        '''Cálculo de retornos de las acciones
        '''
        self.data["log_ret"] = np.log(self.data.Precio/self.data.Precio.shift(1))
        
    def plot_precio(self):
        ''' Gráfico de precios
        '''
        self.data.Precio.plot(figsize = (12, 8))
        plt.title("Gráfico Precios: {}".format(self.ticker), fontsize = 15)
    
    def plot_ret(self, kind = "ts"):
        ''' Gráfico de rentabilidades dos opciones: como serie de tiempo ("ts") o histograma ("hist")
        '''
        if kind == "ts":
            self.data.log_ret.plot(figsize = (12, 8))
            plt.title("Rentabilidad: {}".format(self.ticker), fontsize = 15)
        elif kind == "hist":
            self.data.log_ret.hist(figsize = (12, 8), bins = int(np.sqrt(len(self.data))))
            plt.title("Frecuencia de rentabilidades: {}".format(self.ticker), fontsize = 15)
    
    def set_ticker(self, ticker = None):
        '''Selección nuevo ticker (símbolo)
        '''
        if ticker is not None:
            self.ticker = ticker
            self.get_data()
            self.log_ret()
            
    def mean_ret(self, freq = None):
        '''Cálculo de media de las rentabilidades
        '''
        if freq is None:
            return self.data.log_ret.mean()
        else:
            resampled_precio = self.data.Precio.resample(freq).last()
            resampled_ret = np.log(resampled_precio / resampled_precio.shift(1))
            return resampled_ret.mean()
    
    def std_ret(self, freq = None):
        '''cálculo de desviación típica de las rentabilidades (riesgo)
        '''
        if freq is None:
            return self.data.log_ret.std()
        else:
            resampled_precio = self.data.Precio.resample(freq).last()
            resampled_ret = np.log(resampled_precio / resampled_precio.shift(1))
            return resampled_ret.std()
        
    def anual_perf(self):
        '''cálculo anualizado de rentabilidades y riesgo
        '''
        mean_ret = round(self.data.log_ret.mean() * 252, 3)
        riesgo = round(self.data.log_ret.std() * np.sqrt(252), 3)
        print("Rentabilidad: {} | Riesgo: {}".format(mean_ret, riesgo))
```

```python
stock = AnalisisFinanciero(ticker = "IBE.MC", start = "2015-01-01", 
                            end =  "2019-12-31" )
```

```python
stock.plot_precio()
```

```python
stock.plot_ret()
```

```python
stock.plot_ret(kind = "hist")
```

```python
stock.anual_perf()
```
